```python
from sklearn.preprocessing import StandardScaler
# Veri biliminde bir Kukla Değişkeni (Dummy Variable) standartlaştırmak, 
# onun tüm anlamını yitirmesine neden olur.
# Eğer bir kukla değişken 0 ve 1 olarak kalsaydı, 
# "Bu neden varsa devamsızlık olasılığı 7 kat artar" diyebilirdik.
# Ama onu standartlaştırdığımızda $-0.5$ ve $1.7$ gibi sayılara dönüştüğü için "1 birimlik artış" artık gerçek hayatta hiçbir şeye karşılık gelmez hale geldi.
# 1. Scaler nesnesini tanımla
absenteeism_scaler = StandardScaler()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler
import numpy as np

class CustomScaler(BaseEstimator, TransformerMixin): 
    
    # parametrelerle __init__ metodu
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        # StandardScaler'a bu parametreleri paslıyoruz
        self.scaler = StandardScaler(copy=copy, with_mean=with_mean, with_std=with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None

    # Eğitme metodu, fit metodu 'Future warnings'! ten sonra böyle değişmiştir:
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns], axis=0) # axis=0 eklendi
        self.var_ = np.var(X[self.columns], axis=0)   # axis=0 eklendi
        return self

    # Dönüştürme metodu
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        # Sadece seçtiğimiz sütunları ölçeklendiriyoruz
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        # Seçilmeyen (Dummy) sütunları olduğu gibi bırakıyoruz
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        # Her iki parçayı birleştirip orijinal sütun sırasına diziyoruz
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

## train & test

```python
# 1. Gerekli kütüphaneyi içe aktarıyoruz
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# 2. Veriyi bölüyoruz
# inputs: scaled_inputs (700, 14)
# targets: targets (700,)
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, 
                                                    train_size = 0.8, 
                                                    random_state = 20, 
                                                    shuffle = True)

# 3. Şekilleri (Shapes) kontrol edelim
print("Eğitim Girdileri:", x_train.shape, "Eğitim Hedefleri:", y_train.shape)
print("Test Girdileri:", x_test.shape, "Test Hedefleri:", y_test.shape)
```

```python
# 1. Gerekli kütüphaneleri içe aktaralım
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# 2. Modeli tanımlayalım (Sadece boş bir nesne oluşturuyoruz)
reg = LogisticRegression()

# 3. Modeli eğitelim (Fit işlemi)
# İşte burada modelimiz x_train'deki ipuçları ile y_train'deki sonuçlar arasındaki ilişkiyi öğreniyor.
reg.fit(x_train, y_train)

# 4. Modelin eğitim verisi üzerindeki skorunu (doğruluk oranını) görelim
reg.score(x_train, y_train)
```

```python
# Modelin otomatik skorunu alalım
accuracy_score = reg.score(x_train, y_train)
print(f"Model Doğruluk Oranı: {accuracy_score}")
```

```python
import numpy as np

# 1. Tahminleri alıyoruz
model_outputs = reg.predict(x_train)

# 2. Tahminler ile gerçekleri karşılaştırıyoruz (True/False dizisi)
comparisons = (model_outputs == y_train)

# 3. Manuel skoru hesaplıyoruz (Doğru sayısını toplam sayıya bölüyoruz)
manual_accuracy = np.sum(comparisons) / comparisons.shape[0]

print(f"Manuel Hesaplanan Doğruluk: {manual_accuracy}")
```

```python
import pandas as pd
import numpy as np

# Tabloyu her seferinde sıfırdan oluştur (İndeks kaymasını önler)
# unscaled_inputs.columns.values ifadesi bize 14 değişkenin adını verir.
feature_name = unscaled_inputs.columns.values
summary_table = pd.DataFrame(columns=['Feature name'], data=feature_name)

# Katsayıları (Weights) ekle
# reg.coef_ (1, 14) boyutunda olduğu için transpose (devriğini) alıyoruz.
summary_table['Coefficient'] = np.transpose(reg.coef_)

# İndeksi manipüle etmeden önce temiz bir kopyasını al
# Bu satır, hücreyi 100 kere de çalıştırsan indeksin hep 1'den başlamasını sağlar
summary_table.index = summary_table.index + 1

# Intercept (Bias) değerini en başa (0. indekse) ekle
# reg.intercept_[0] bize o tekil sabit sayıyı verir.
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]

# Tabloyu indekse göre sırala ve Odds Ratio hesapla
summary_table = summary_table.sort_index()
summary_table['Odds ratio'] = np.exp(summary_table.Coefficient)

# Kusursuz 15 satırlık (14 feature + 1 intercept) özet tablo
summary_table
```
